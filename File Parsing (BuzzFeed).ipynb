# Parse multiple xml into a dataframe

```python
from os import listdir, path
import xml.etree.ElementTree as ET
import codecs,os
import pandas as pd



```

```python
mypath = '../Thesis/BuzzFeed-Webis'

files = [path.join(mypath, f) for f in listdir(mypath) if f.endswith('.xml')]
doc_id = []
mainText = [] 
orientation = []
veracity = []
title = []
author = []
portal = []

for file in files:   
    print(file)
    tree = ET.parse(file)
    root = tree.getroot()
    doc_number = ET.Element("number")
    doc_number.text = file[25:29]
    root.append(doc_number)
    content = root.find("mainText").text
    doc_number = root.find("number").text
    orien = root.find("orientation").text
    vera = root.find("veracity").text
    NewsTitle = root.find("title").text
    NewsAuthor = root.find("author").text
    NewsPortal = root.find("portal").text
    doc_id.append(doc_number)
    mainText.append(content)
    orientation.append(orien)
    veracity.append(vera)
    title.append(NewsTitle)
    author.append(NewsAuthor)
    portal.append(NewsPortal)
        
data = pd.DataFrame({'doc_id': doc_id, 'author': author, 'title': title, 'mainText': mainText, 'portal': portal, 'orientation': orientation, 'veracity': veracity})

    
#    output = codecs.open(file,"w","utf-8")
#    output.write(content)
#    output.close()
    
```
