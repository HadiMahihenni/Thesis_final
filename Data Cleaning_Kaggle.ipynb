```python
import numpy as np
import pandas as pd
```

# Load dataset

```python
df = pd.read_csv('Datasets/Kaggle/fake.csv')
df.head()
```

```python
df.shape
```

# Assign ID

```python
# set doc_id
df.insert(0, 'doc_id', range(0, len(df)))
```

```python
len(df)
```

# Data Cleaning

```python
#delete unnecessary columns
df = df[df.language == 'english']
df = df[df.country == 'US']
```

```python
len(df)
```

```python
#delete unnecessary columns
df = df[['doc_id','author','published','title', 'text', 'thread_title', 'type']]
```

```python
#label data as fake
df['type'] = False
```

```python
df.head()
```

```python
# keep only data which is published on 26.10 and 27.19
df['published'] = df['published'].str[0:10]
array = ['2016-10-26', '2016-10-27']
df = df.loc[df['published'].isin(array)]
len(df)
```

```python
df['word_count'] = df['text'].apply(lambda x: len(str(x).split(" ")))
df1 = df.drop_duplicates(subset=['title', 'text'])
df1
```

```python
# Keep data only the content has more than 200 words
df2 = df1[df1['word_count'] > 200]  
df3 = df2[pd.notnull(df2['title'])]
```

```python
df3 = df3.reset_index(drop=True)
df3.isnull().sum()
```

```python
# gerneral preproceesing and delete missing value
df3.text = df.text.str.strip().str.replace('(', '').str.replace(')', '').str.replace("Before It's News\n","").str.replace("\n","").str.replace("Oct","").str.replace("PM","")
df4 = df3.dropna(subset=['text', 'author'])
df5 = df4.drop_duplicates(subset=['text'])
df5.isnull().sum()
```

```python
len(df5)
```

```python
#delete unnecessary columns
df_extracted = df5[['doc_id','title','text','type']]
df_extracted.columns = ['doc_id','title','mainText','veracity']
```

```python
df_extracted.head()
```

```python
df_extracted = df_extracted.reset_index(drop=True)
df_extracted["doc_id"] = df_extracted.index
```

```python
df_extracted.to_csv("Kaggle.csv",index=False)
```
